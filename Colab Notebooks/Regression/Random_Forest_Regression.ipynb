```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
df=pd.read_csv('/content/drive/MyDrive/Machine Learning/Part 2 - Regression/Section 9 - Random Forest Regression/Python/Position_Salaries.csv')
X=df.iloc[:,1:-1].values
Y=df.iloc[:,-1].values
```

## Training the Random Forest Regression model on the whole dataset

```
from sklearn.ensemble import RandomForestRegressor
rr=RandomForestRegressor(n_estimators=10,random_state=0)
rr.fit(X,Y)
```

## Predicting a new result

```
rr.predict([[6.5]])
```

## Visualising the Random Forest Regression results (higher resolution)

```
X_grid=np.arange(min(X),max(X),0.1)
X_grid=X_grid.reshape((len(X_grid),1))
plt.scatter(X,Y,color='red')
plt.plot(X_grid,rr.predict(X_grid),color='blue')
plt.title('Truth or Bluff(Random Forest Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
