```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/Machine Learning/Part 2 - Regression/Section 5 - Multiple Linear Regression/Python/50_Startups.csv')
X=df.iloc[:,:-1].values
Y=df.iloc[:,-1].values
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
print(X)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(X_train,Y_train)
```

```python
Y_pred=lr.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((Y_pred.reshape(len(Y_pred),1),Y_pred.reshape(len(Y_test),1)),1))
```
